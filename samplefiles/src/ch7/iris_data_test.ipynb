```python
import random, numpy as np

# 二つのリストを合成
d = list(zip(iris.data.tolist(), iris.target))

# シャッフル
random.shuffle(d)

# 8:2の割合で分ける
train_data = d[0:130]
test_data = d[130:150]

# 合成したリストを分離する
x_train, y_train = zip(*train_data)
x_test, y_test = zip(*test_data)
np.array(x_train)
```

```python
from sklearn import svm
# モデルを生成
clf = svm.SVC()
# データを学習
clf.fit(x_train, y_train)
```

```python
# テストデータを利用して予測
pred = clf.predict(x_test)
# 予測が答えと、どれくらい合っているかカウント
result = list(pred == y_test).count(True) / len(y_test)
print("正解率=" + str(result))
```

```python
cnt = 0
for pre, y in zip(pred, y_test):
    result = (pre == y)
    print(pre, y, result)
    if result: cnt += 1
print("{0}/{1}={2}".format(cnt, len(y_test), cnt / len(y_test)))
```
